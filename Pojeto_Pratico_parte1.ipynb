- Contabilize os motivos de devolução por cliente
    - Considerar apenas os pedidos com devolução, seguindo os seguintes passos:
        - filtrar apenas os dados de pedidos com devolução
        - agrupar o resultado por cliente e motivo da devolução
        - realizar a contagem pelo número de pedido
    - Verificar qual foi o maior motivo de devolução dos pedidos.

```python
# Primeiramente vamos ver os status dos pedidos

dados["motivo_devolucao"].value_counts()
```

```python
filtro = dados["motivo_devolucao"] != "S/ Devolu."  # Com essa linha excluimos o "S/ Ddevolução", apresentando apenas os demais motivos
dados_com_devolucao = dados.loc[filtro]
```

```python
colunas = ["cliente", "motivo_devolucao"]           # apresentação do cliente e o motivo da devoleção
dados_com_devolucao.groupby(colunas)["numero_pedido"].count().to_frame()
```

- Acima conseguimos filtrar e perceber onde estão ocorrendo erros na distribuição dos produtos, e o maior erro está no envio de produtos errados.

- Criando um relatório com o valor total do prejuízo causado por cada motivo de devolução, agrupado por cliente.
    - Considerar apenas os pedidos com devolução!
        - Exportar esse relatório para um arquivo Excel!


```python
dados_com_devolucao
```

```python
# Agrupando os dados solicitados

colunas = ["cliente","motivo_devolucao"]
dados_com_devolucao.groupby(colunas)["valor_pedido"].sum().to_excel("Prejuízo por motivo de devolução.xlsx")
```

- Será com a base de dados acima que iremos trabalhar

```python
filtro = dados["motivo_devolucao"] == "S/ Devolu."
dados_sem_devolucao = dados.loc[filtro]
```

```python
dados_sem_devolucao ["status"].value_counts()
```

- Acima os pedidos que NÃO tiveram devolução, os que foram certos. Porém nos revela que, a grande maioria destes produtos chegou com atrasado.

```python
# Agrupando nossas informações

colunas = ["cliente", "motorista", "status"]
dados_sem_devolucao.groupby(colunas)["numero_pedido"].count().to_frame()
```

```python
# Exportando os dados para o Excel

colunas = ["cliente", "motorista", "status"]
dados_sem_devolucao.groupby(colunas)["numero_pedido"].count().to_excel("Status de entrega por motoristas e clientes.xlsx")
```
